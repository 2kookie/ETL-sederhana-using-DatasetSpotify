```python
spotify.dropna(how="all", inplace=True)
spotify.isna().sum()
```

```python
# Identifikasi outlier dalam kolom tertentu (misalnya, danceability)
column_name = 'danceability'
Q1 = spotify[column_name].quantile(0.25)
Q3 = spotify[column_name].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menghapus baris dengan outlier
df_cleaned = spotify[(spotify[column_name] >= lower_bound) & (spotify[column_name] <= upper_bound)]
```

```python
import pandas as pd

# Load the DataFrame
df_cleaned = pd.read_csv("Spotify_Song.csv")

# Group the data by genres and create a dictionary of artists for each genre
genre_artists = df_cleaned.groupby('genre')['artistName'].unique().apply(list).to_dict()

# Print the artists for each genre
for genre, artists in genre_artists.items():
    print(f"Genre: {genre}")
    print("Artists:", ", ".join(artists))
    print("-" * 50)
```

MENGKATEGORIKAN GENRE MUSIK DAN ARTIS BERDASARKAN POPULARITAS


```python


# Calculate the total plays for each genre
genre_popularity = df_cleaned.groupby('genre')['msPlayed'].sum().reset_index()

# Sort genres by popularity in descending order
genre_popularity = genre_popularity.sort_values(by='msPlayed', ascending=False)


# Plot the top 10 genres by popularity
plt.figure(figsize=(10, 6))
plt.bar(genre_popularity['genre'].head(10), genre_popularity['msPlayed'].head(10), color = "green")
plt.xlabel('Genre')
plt.ylabel('Total Plays')
plt.title('Top 10 Music Genres by Popularity')
plt.xticks(rotation=45)
plt.show()

# Calculate the total plays for each artist
artist_popularity = df_cleaned.groupby('artistName')['msPlayed'].sum().reset_index()

# Sort artists by popularity in descending order
artist_popularity = artist_popularity.sort_values(by='msPlayed', ascending=False)

# Plot the top 10 artists by popularity
plt.figure(figsize=(12, 6))
plt.bar(artist_popularity['artistName'].head(10), artist_popularity['msPlayed'].head(10), color='black')
plt.xlabel('Artist')
plt.ylabel('Total Plays')
plt.title('Top 10 Artists by Popularity')
plt.xticks(rotation=45)
plt.show()

weight_popularity = 0.4
weight_danceability = 0.2
weight_energy = 0.2
weight_valence = 0.2

# Calculate a score for each track based on the criteria
df_cleaned['score'] = (
    weight_popularity * df_cleaned['msPlayed'] +
    weight_danceability * df_cleaned['danceability'] +
    weight_energy * df_cleaned['energy'] +
    weight_valence * df_cleaned['valence']
)

# Sort tracks by score in descending order to find the best music/songs
best_tracks = df_cleaned.sort_values(by='score', ascending=False).head(20)

# Find the best artists based on the total score of their songs
best_artists = df_cleaned.groupby('artistName')['score'].sum().reset_index()
best_artists = best_artists.sort_values(by='score', ascending=False).head(10)

# Visualize the best music/songs using a horizontal bar chart
plt.figure(figsize=(10, 4))
plt.barh(best_tracks['trackName'], best_tracks['score'], color='lightgreen')
plt.xlabel('Score')
plt.ylabel('Track Name')
plt.title('Top 10 Best Music/Songs')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

best_tracks.head(10)
```

```python

from textblob import TextBlob
```

```python
# Function to perform sentiment analysis on track names
def get_sentiment(track_name):
    blob = TextBlob(track_name)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score > 0:
        return 'positive'
    elif sentiment_score < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to track names
df_cleaned['sentiment'] = df_cleaned['trackName'].apply(get_sentiment)
```

```python
# Pie chart for sentiment distribution by genre
genre_sentiment_counts = df_cleaned.groupby(['genre', 'sentiment']).size().unstack(fill_value=0)
genre_sentiment_counts.plot(kind='pie', subplots=True, figsize=(15, 5), autopct='%1.1f%%', legend=False)
plt.title('Sentiment Distribution by Genre')
plt.show()
```

```python
# Pie chart for sentiment distribution by artist
artist_sentiment_counts = df_cleaned.groupby(['artistName', 'sentiment']).size().unstack(fill_value=0)
artist_sentiment_counts['total'] = artist_sentiment_counts.sum(axis=1)
top_10_artists = artist_sentiment_counts.nlargest(10, 'total').drop(columns='total')
top_10_artists.plot(kind='pie', subplots=True, figsize=(15, 5), autopct='%1.1f%%')
plt.title('Sentiment Distribution of Top 10 Artists')
plt.show()
```
